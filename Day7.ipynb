```python
'''function to take three students names
 and marks using keyword arguments
 and print the name and marks of
the student with highest marks'''
def highest_marks(**students):
    highest_marks = 0
    highest_marks_student = None

    for name, marks in students.items():
        if marks > highest_marks:
            highest_marks = marks
            highest_marks_student = name

    print(f"Student with the highest marks is {highest_marks_student} with {highest_marks} marks.")
highest_marks(Student1=80, Student2=90, Student3=85)
```

```python
#A function to return domain name from a website name

def get_domain_name(website_url):

    parts = website_url.split('.')

    domain_name = parts[-1]

    return domain_name

website_url = "https://www.example.com"
domain_name = get_domain_name(website_url)
print("Domain name:", domain_name)
```

```python

'''program to take month and year
 and calculate age in years and
months up to march-2024'''
from datetime import datetime
def calculate_age(month, year):
    current_date = datetime.now()
    march_2024 = datetime(2024, 3, 1)
    dob = datetime(year, month, 1)
    age = march_2024 - dob
    age_years = age.days // 365
    age_months = (age.days % 365) // 30

    return age_years, age_months

month = int(input("Enter the birth month (1-12): "))
year = int(input("Enter the birth year: "))
age_years, age_months = calculate_age(month, year)
print(f"Age up to March 2024: {age_years} years and {age_months} months")
```

```python
#Write a function to take 4-bit binary number and check whether odd or even.
def is_even_or_odd(binary):
    decimal = int(binary, 2)
    if decimal % 2 == 0:
        return "Even"
    else:
        return "Odd"
binary_number = input("Enter a 4-bit binary number: ")
if len(binary_number) != 4 or not all(bit in '01' for bit in binary_number):
    print("Invalid input! Please enter a valid 4-bit binary number.")
else:
    result = is_even_or_odd(binary_number)
    print(f"The binary number {binary_number} is {result}.")
```

```python
# write a function to return average of a list.
def avg(nums):
    return sum(nums)/len(nums)
mylist = [92,91,71,52,81,72,55,67,76]
print('My_list:',mylist)
print('Averge of list:',avg(mylist))
```

```python
'''
Write a function to take sequence of number separated by comma and return list of numbers.
Eg: fun (“2,4,1,5,3”) returns list = [2,4,1,5,3]
'''
def num_list(string):
    nums = [int(i) for i in string.split(',')]
    return nums
string = input("Enter sequence of numbers separated by comma's:")
print('Numbers list:',num_list(string))
```
