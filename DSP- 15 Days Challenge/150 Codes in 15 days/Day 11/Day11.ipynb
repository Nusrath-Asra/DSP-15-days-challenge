```python
# Write a function to return amount of electricity bill by taking no of units consumed.
def calculate_electricity_bill(units):

  if units <= 100:
    bill = units * 4.5
  elif units <= 200:
    bill = 100 * 4.5 + (units - 100) * 6
  elif units <= 300:
    bill = 100 * 4.5 + 100 * 6 + (units - 200) * 10
  else:
    bill = 100 * 4.5 + 100 * 6 + 100 * 10 + (units - 300) * 20

  return bill
units_consumed = 250
total_bill = calculate_electricity_bill(units_consumed)
print(f"Total electricity bill for {units_consumed} units: {total_bill:.2f}")
```

```python
#program to print the list of students who have pass
def check_pass(student_marks):
  passed_students = []
  for student, marks in student_marks.items():
    internal_marks, external_marks = marks
    if internal_marks >= 30 and external_marks > 20:
      passed_students.append(student)

  return passed_students
student_marks = {
  "Alice": [45, 25],
  "Bob": [20, 35],
  "Charlie": [30, 40],
  "David": [15, 45],
}

passed_students = check_pass(student_marks)

print("List of students who have passed:")
for student in passed_students:
  print(f"\t- {student}")
```

```python
'''a debate there are two team - positive, negative:
Ten applicants ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
are shortlisted for the interview.
randomly select five applicants for each team'''
import random

applicants = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
random.shuffle(applicants)
team_positive = applicants[:5]
team_negative = applicants[5:]

print("Team Positive:", team_positive)
print("Team Negative:", team_negative)
```

```python
factorial = lambda n: 1 if n == 0 else n * factorial(n-1)

def calculate_factorial(n):
  if n < 0:
    raise ValueError("Factorial is not defined for negative numbers.")
  return factorial(n)
print(calculate_factorial(5))
```

```python
import re

def remove_special_characters(string):

  pattern = r"[^\w\s]"
  cleaned_string = re.sub(pattern, "", string)

  return cleaned_string

string_with_special_characters = "!@#$$%^&*()_+"
cleaned_string = remove_special_characters(string_with_special_characters)

print("Original string:", string_with_special_characters)
print("Cleaned string:", cleaned_string)
```

```python

def arrange_characters_by_ascii(string):

  ascii_values = [ord(char) for char in string]

  ascii_values.sort()
  arranged_string = ''.join([chr(value) for value in ascii_values])

  return arranged_string
string = "bcadef"
arranged_string = arrange_characters_by_ascii(string)

print("Original string:", string)
print("Arranged string:", arranged_string)
```

```python
#7. Write a program to find difference(subtraction) between maximum and minimum element of list.
def find_difference_max_min(list1):
  max_element = max(list1)
  min_element = min(list1)
  difference = max_element - min_element

  return difference
list1 = [10, 20, 4, 5, 6, 7, 8, 9]
difference = find_difference_max_min(list1)
print("Difference between maximum and minimum element:", difference)
```

```python
#a function to take list of two-digit numbers and round- off to nearest ten.
def round_off_to_nearest_ten(list_of_numbers):
  rounded_numbers = []

  for number in list_of_numbers:
    remainder = number % 10
    if remainder >= 5:
      rounded_number = number + (10 - remainder)
    else:
      rounded_number = number - remainder

    rounded_numbers.append(rounded_number)

  return rounded_numbers
list_of_numbers = [12, 15, 23, 27, 34, 38, 45, 49]
rounded_numbers = round_off_to_nearest_ten(list_of_numbers)
print("Rounded-off numbers:", rounded_numbers)
```

```python
#Write a function to take 2-D matrix as parameter and return determinant of matrix
import numpy as np

def calculate_determinant(matrix):
  if matrix.shape[0] != matrix.shape[1]:
    raise ValueError("Matrix must be square to calculate determinant.")
  determinant = np.linalg.det(matrix)

  return determinant

matrix = np.array([[1, 2], [3, 4]])

determinant = calculate_determinant(matrix)
print("Determinant of the matrix:", determinant)

```

```python
#program to find greatest common divisor of two numbers.
def gcd(a, b):
  while b != 0:
    a, b = b, a % b

  return a

num1 = 24
num2 = 36
gcd_value = gcd(num1, num2)
print("The GCD of", num1, "and", num2, "is:", gcd_value)
```
