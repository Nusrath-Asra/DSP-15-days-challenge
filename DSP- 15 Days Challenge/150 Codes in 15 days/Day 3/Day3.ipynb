```python

#to calculate the sum of even and odd
def fun():
    even=0
    odd=0
    u=int(input("Upper limit"))
    l=int(input("Lower limit"))
    for num in range(l,u):
      if num%2 ==0:
        even+=num
      else:
        odd+=num
    return even,odd
even,odd=fun()
print("sum of even b/w 50: ",even)
print("sum of odd b/w 50: ",odd)
```
