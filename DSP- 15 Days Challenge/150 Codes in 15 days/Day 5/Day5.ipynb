```python
#to check if string is binary or not
string = input("Enter a string: ")
is_binary = True
for char in string:
    if char != '0' and char != '1':
        is_binary = False
        break
if is_binary:
    print("The string is binary.")
else:
    print("The string is not binary.")
```

```python
#to find common words from 2 strings
s1=input("String 1:")
s2=input("String 2:")
string1=set(s1.split())
string2=set(s2.split())
common=string1.intersection(string2)
print(common)
```

```python
s1=input("String 1:").split()
s2=input("String 2:").split()
common=[]
for word in s1:
  for words in s2:
    if word==words:
      common.append(word)
print(common)
```
