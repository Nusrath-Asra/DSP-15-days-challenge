```python

#if num is composite or not
import math
n=int(input("Enter: "))
if n > 1:
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            print(f"{n} is composite.")
            break
    else:
        print(f"{n} is not composite.")
else:
    print(f"{n} is not composite.")
```

```python

#to find first to sum of n natural numbers
total = 0
n=int(input("Num: "))
for i in range(1, n + 1):
    total += i**2
print("Required: ", n, "natural numbers:", total)
```

```python
#to check if armstrong or not
number = 153
num_str = str(number)
sum_of_digits = 0

for digit in num_str:
    digit_power = int(digit) ** 3
    sum_of_digits += digit_power

if sum_of_digits == number:
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```
