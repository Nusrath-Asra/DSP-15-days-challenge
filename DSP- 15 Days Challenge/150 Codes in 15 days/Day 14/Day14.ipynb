```python
#program to find roots of quadratic equation.
import cmath
def quadratic_roots(a, b, c):

    discriminant = (b**2) - (4*a*c)

    # Find two solutions
    root1 = (-b - cmath.sqrt(discriminant)) / (2*a)
    root2 = (-b + cmath.sqrt(discriminant)) / (2*a)

    return root1, root2
a = 1
b = -3
c = 2

root1, root2 = quadratic_roots(a, b, c)
print("Roots of the quadratic equation:")
print("Root 1:", root1)
print("Root 2:", root2)
```

```python
#program to transpose 3*3 matrix
def transpose_matrix(matrix):
    transposed_matrix = []
    for i in range(len(matrix[0])):
        transposed_row = []
        for row in matrix:
            transposed_row.append(row[i])
        transposed_matrix.append(transposed_row)
    return transposed_matrix

matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

transposed_matrix = transpose_matrix(matrix)
print("Original Matrix:")
for row in matrix:
    print(row)
print("\nTransposed Matrix:")
for row in transposed_matrix:
    print(row)
```

```python
#check whether a list follow ascending or descending or no- order.
def check_order(lst):
    is_ascending = all(lst[i] <= lst[i+1] for i in range(len(lst)-1))
    is_descending = all(lst[i] >= lst[i+1] for i in range(len(lst)-1))

    if is_ascending:
        return "Ascending order"
    elif is_descending:
        return "Descending order"
    else:
        return "No particular order"

# Example usage:
ascending_list = [1, 2, 3, 4, 5]
descending_list = [5, 4, 3, 2, 1]
unordered_list = [3, 1, 5, 2, 4]

print("List:", ascending_list, "-", check_order(ascending_list))
print("List:", descending_list, "-", check_order(descending_list))
print("List:", unordered_list, "-", check_order(unordered_list))
```

```python
#program to multiply a column matrix with row matrix.
def multiply_matrices(col_matrix, row_matrix):
    result = 0
    for i in range(len(col_matrix)):
        result += col_matrix[i] * row_matrix[i][0]
    return result
col_matrix = [1, 2, 3]
row_matrix = [[4],
              [5],
              [6]]

result = multiply_matrices(col_matrix, row_matrix)
print("Result of multiplying column matrix with row matrix:", result)
```

```python
'''playlist contains five songs.
 write a program to shuffle the playlist n-times which is on
 loop for n times taken from user.'''
import random

def shuffle_playlist(playlist, shuffles, repetitions):
    for _ in range(repetitions):
        for _ in range(shuffles):
            random.shuffle(playlist)
        print("Shuffled playlist:", playlist)

playlist = ['Song 1', 'Song 2', 'Song 3', 'Song 4', 'Song 5']
n_shuffles = int(input("Enter the number of shuffles: "))
n_repetitions = int(input("Enter the number of repetitions: "))

shuffle_playlist(playlist, n_shuffles, n_repetitions)
```

```python
'''write a program to remove
 duplicate characters from string irrespective
 of case and print in same order.'''
def remove_duplicates(string):
    seen = set()
    result = []
    for char in string:
        if char.lower() not in seen:
            seen.add(char.lower())
            result.append(char)
    return ''.join(result)

input_string = "Programming"
output_string = remove_duplicates(input_string)
print("String after removing duplicates:", output_string)
```

```python
#write a program to calculate geometric mean on list.
def geometric_mean(lst):
    if not lst:
        return None
    product = 1
    for num in lst:
        product *= num
    return product**(1/len(lst))

numbers = [2, 4, 8, 16]
result = geometric_mean(numbers)
print("Geometric mean of the numbers:", result)
```

```python
#write a program to calculate compound interest.
def compound_interest(principal, rate, time):
    amount = principal * (1 + rate / 100) ** time
    compound_interest = amount - principal
    return compound_interest

principal_amount = float(input("Enter the principal amount: "))
annual_interest_rate = float(input("Enter the annual interest rate (in percentage): "))
time_period = float(input("Enter the time period (in years): "))

interest = compound_interest(principal_amount, annual_interest_rate, time_period)
print("Compound interest:", round(interest, 2))
```

```python
#list contains both strings and numbers concatenate all the elements into a string.
def concatenate_elements(my_list):
    concatenated_string = ''.join(map(str, my_list))
    return concatenated_string

my_list = ['Hello', 123, 'World', 456]
result = concatenate_elements(my_list)
print("Concatenated string:", result)
```

```python
#program to import counter from collections:
from collections import Counter

def most_frequent_element(lst):
    count = Counter(lst)
    most_common = count.most_common(1)
    return most_common[0][0] if most_common else None

def least_frequent_element(lst):
    count = Counter(lst)
    least_common = count.most_common()[-1]
    return least_common[0] if least_common else None

def display_elements_with_counter(lst):
    count = Counter(lst)
    print("Elements of the list with their counts:")
    for element, frequency in count.items():
        print(f"{element}: {frequency}")

my_list = [1, 2, 3, 2, 2, 4, 5, 5, 5]
most_frequent = most_frequent_element(my_list)
least_frequent = least_frequent_element(my_list)

print("Most frequent element:", most_frequent)
print("Least frequent element:", least_frequent)

display_elements_with_counter(my_list)
```
