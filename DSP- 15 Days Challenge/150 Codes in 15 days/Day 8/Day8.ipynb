```python

'''A list contains both odd and
even numbers. Write a function
that takes list and return dictionary
where odd and even numbers are stored
 separately.'''
def fun(numbers):
  dic={'even':[],'odd':[]}
  for i in numbers:
    if i%2==0:
      dic['even'].append(i)
    else:
      dic['odd'].append(i)
  return dic
dictionary=fun([23,12,45,64,22,48,87,90,1,2,3])
print(dictionary)
```

```python
#program to check palindrome or not using recursion.
def is_palindrome(s):
    # Base case: if the string is empty or has only one character, it's a palindrome
    if len(s) <= 1:
        return True


    if s[0] == s[-1]:

        return is_palindrome(s[1:-1])
    else:
        # If they don't match, it's not a palindrome
        return False

# Test the function
word = input("Enter a word to check if it's a palindrome: ")
if is_palindrome(word):
    print(f"{word} is a palindrome.")
else:
    print(f"{word} is not a palindrome.")
```

```python
'''A dictionary contains three
genres and three movies of each genre
a function to take gerne and return
 a random movie of that genre'''
import random
dic={'Mystery':['The Da Vinci Code','Enola Holmes','Zodiac'],
     'Novel_based':['Pride & prejudice','Fault in our stars','Little Women'],
     'Action':['Resident Evil','Inception','John Wick']}
def fun(g):
    if g in dic:
        movies_list = dic[g]
        return random.choice(movies_list)
    else:
        return "Genre not found in the dictionary."

g = input("Enter a genre (action, novel_based, Mystery): ")
random_movie = fun(g)
print("Random movie from the genre",random_movie)
```

```python
'''A list contains valid and invalid
 Gamils of students registering course
 in order. Write a program to select
 3 students with valid Gmail using
 first-come-first-serve.'''
def select_students(emails):
    valid_students = []
    for email in emails:
        if is_valid_gmail(email):
            valid_students.append(email)
            if len(valid_students) == 3:
                break
    return valid_students
def is_valid_gmail(email):
    if email.endswith('@gmail.com') and len(email) > len('@gmail.com'):
        return True
    return False
emails = ['asragmail.com', 'nus@hotmail.com', 'ahmed@gmail.com', 'mohammad@yahoo.com', 'nussu@gmail.com']
selected_students = select_students(emails)
print("Selected students with valid Gmail addresses (first-come-first-serve):")
for student in selected_students:
    print(student)
```

```python
'''Write a function to take list
and return multiplication of elements
 of list.'''
def multiply_list(lst):
    result = 1
    for num in lst:
        result *= num
    return result

my_list = [2, 3, 4, 5]
print("Multiplication of elements in the list:", multiply_list(my_list))
```

```python
'''product has six months warranty.
Write a program to check warranty
of product by taking date month of
purchase and calculate up to 25 - March.'''

def check_warranty(purchase_date, purchase_month):
    thirty_one_days_months = {1, 3, 5, 7, 8, 10, 12}
    thirty_days_months = {4, 6, 9, 11}

    warranty_months = 6

    remaining_months = 0
    if purchase_month in thirty_one_days_months:
        remaining_months = 31 - purchase_date
    elif purchase_month in thirty_days_months:
        remaining_months = 30 - purchase_date
    else:
        remaining_months = 28 - purchase_date

    if remaining_months < warranty_months:
        print("Warranty has expired.")
    else:
        print("Warranty is still valid.")

purchase_date = int(input("Enter the date of purchase: "))
purchase_month = int(input("Enter the month of purchase: "))

check_warranty(purchase_date, purchase_month)
```

```python

'''Print the permutations of
'a', 'b', 'c' without repetition.
Output['abc', 'acb', 'bac', 'bca', 'cab', 'cba']'''
from itertools import permutations
permutations_list = [''.join(perm) for perm in permutations('abc')]

print(permutations_list)
```

```python
'''list contains ingredients of product. Avoid the product
if it has both wheat and milk'''

def avoid_product(ingredients):
    if 'wheat' in ingredients and 'milk' in ingredients:
        print("Avoid the product because it contains both wheat and milk.")
    else:
        print("You can consume the product.")

product_1 = ['sugar', 'salt', 'wheat']
product_2 = ['sugar', 'salt', 'milk']
product_3 = ['sugar', 'salt', 'wheat', 'milk']

avoid_product(product_1)
avoid_product(product_2)
avoid_product(product_3)
```

```python

#function to calculate simple interest
def calculate_simple_interest(principal, rate, time):
    simple_interest = (principal * rate * time) / 100
    return simple_interest

principal_amount = float(input("Enter the principal amount: "))
interest_rate = float(input("Enter the interest rate (in percentage): "))
time_period = float(input("Enter the time period (in years): "))

interest = calculate_simple_interest(principal_amount, interest_rate, time_period)
print("Simple interest:", interest)
```
