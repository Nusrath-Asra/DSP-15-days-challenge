```python
#Write a program to calculate the sum of list by index_range taken from console.
start_index = 0
end_index=0
def sum_list_by_index_range(list_of_numbers):
  start_index = int(input("Enter the starting index: "))
  end_index = int(input("Enter the ending index: "))

  if start_index < 0 or end_index >= len(list_of_numbers) or start_index > end_index:
    raise ValueError("Invalid index range.")

  sum_of_elements = 0
  for i in range(start_index, end_index + 1):
    sum_of_elements += list_of_numbers[i]

  return sum_of_elements
list_of_numbers = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]
sum_of_elements = sum_list_by_index_range(list_of_numbers)
print(f"The sum of the elements from index {start_index} to {end_index} is: {sum_of_elements}")
```

```python
#Write a function to return whether an 8-bit binary string has equal no of zeros and ones or not (balanced or unbalanced).
def is_binary_string_balanced(binary_string):

  if not binary_string or not all(c in ['0', '1'] for c in binary_string):
    raise ValueError("Invalid binary string.")
  num_zeros = binary_string.count('0')
  num_ones = binary_string.count('1')

  return num_zeros == num_ones

binary_string = "00011100"
is_balanced = is_binary_string_balanced(binary_string)

if is_balanced:
  print(f"The binary string '{binary_string}' is balanced.")
else:
  print(f"The binary string '{binary_string}' is unbalanced.")
```

```python
#4. Write a function to take two values as parameters and return ratio of two values.
def calculate_ratio(numerator, denominator):

  if denominator == 0:
    raise ValueError("Denominator cannot be zero.")

  ratio = numerator / denominator

  return ratio

numerator = 10
denominator = 5

ratio = calculate_ratio(numerator, denominator)
print(f"The ratio of {numerator} and {denominator} is: {ratio}")
```

```python
# a function to return dictionary with brand and final price.
def calculate_final_prices(product_dict):

  final_prices = {}

  for brand, values in product_dict.items():
    mrp = values[0]
    discount_percentage = values[1]
    discount_amount = (discount_percentage / 100) * mrp

    final_price = mrp - discount_amount

    final_prices[brand] = final_price

  return final_prices
product = {'brand-1': [100, 10], 'brand-2': [200, 20], 'brand-3': [300, 30]}

final_prices = calculate_final_prices(product)
print(final_prices)
```

```python
# a function to return the list of capital letters in a string.
def get_capital_letters(string):
  capital_letters = []
  for char in string:
    if char.isupper():
      capital_letters.append(char)

  return capital_letters
string = "The quick Brow Fox"

capital_letters = get_capital_letters(string)
print(f"Capital letters in the string '{string}': {capital_letters}")
```

```python
#write a program to display k column of 3*3 matrix

def display_kth_column(matrix, k):
  if k < 0 or k > 2:
    raise ValueError("Invalid column index.")

  for row in matrix:
    print(row[k])

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
k = 1

display_kth_column(matrix, k)
```

```python
# program to take list of integers and convert their signs i.e. positive to negative and vice-versa
def invert_signs(nums):

  inverted_nums = []

  for num in nums:
    inverted_nums.append(-num)

  return inverted_nums

nums = [1, -2, 3, -4, 5]

inverted_nums = invert_signs(nums)
print(f"Original list: {nums}")
print(f"Inverted list: {inverted_nums}")
```

```python
'''Write a function to take shape of two matrix and check whether a matrix multiplication is possible or not.'''
def matrix_multiplication(matrix1, matrix2):
  if len(matrix1[0]) != len(matrix2):
    raise ValueError("Matrix multiplication is not possible.")

  result = [[0 for _ in range(len(matrix2[0]))] for _ in range(len(matrix1))]
  for i in range(len(matrix1)):
    for j in range(len(matrix2[0])):
      for k in range(len(matrix2)):
        result[i][j] += matrix1[i][k] * matrix2[k][j]
  return result

matrix1 = [[1, 2], [3, 4]]
matrix2 = [[5, 6], [7, 8]]

result = matrix_multiplication(matrix1, matrix2)

print("Result of matrix multiplication:")
for row in result:
  print(row)
```
