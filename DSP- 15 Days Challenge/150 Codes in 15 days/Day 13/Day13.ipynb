```python
def rotate_list(lst, index):
    if index >= len(lst):
        print("Index out of range.")
        return
    rotated_list = lst[index:] + lst[:index]
    return rotated_list

my_list = ['a', 'b', 'c', 'd', 'e']
rotation_index = 2
rotated_list = rotate_list(my_list, rotation_index)
print(rotated_list)
```

```python
def sort_list_ignore_sign(lst):
    sorted_list = sorted(lst, key=lambda x: abs(x))
    return sorted_list

my_list = [9, -4, 2, 0, -2, 7, -3, 5, -5]
sorted_list = sort_list_ignore_sign(my_list)
print("Sorted list:", sorted_list)
```

```python
'''Write a function that take a
sentence and returns the sentence
by shuffling the words.'''
import random

def shuffle_sentence(sentence):
    words = sentence.split()
    random.shuffle(words)
    shuffled_sentence = ' '.join(words)
    return shuffled_sentence

input_sentence = "This is a sample sentence to shuffle words."
shuffled_sentence = shuffle_sentence(input_sentence)
print("Original sentence:", input_sentence)
print("Shuffled sentence:", shuffled_sentence)
```

```python
#function to return LCM of two numbers.
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def lcm(a, b):
    return abs(a * b) // gcd(a, b)

num1 = 12
num2 = 15
lcm_result = lcm(num1, num2)
print("LCM of", num1, "and", num2, "is:", lcm_result)
```

```python
'''program to delete the words form list if it contains
more than two vowel characters.'''
def count_vowels(word):
    vowels = 'aeiouAEIOU'
    return sum(1 for char in word if char in vowels)

def filter_words(words):
    return [word for word in words if count_vowels(word) <= 2]

word_list = ['Book', 'mobile', 'queue', 'python', 'apple', 'algorithm']
filtered_list = filter_words(word_list)
print("Filtered list:", filtered_list)
```

```python
'''function to take a n-digits number and check whether it is dominated by even
 number or odd numbers or balanced.'''
def check_dominance(number):
    even_count = 0
    odd_count = 0
    for digit in str(number):
        if int(digit) % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    if even_count > odd_count:
        return "dominated by even numbers"
    elif even_count < odd_count:
        return "dominated by odd numbers"
    else:
        return "balanced"
num1 = 1234
num2 = 2234
num3 = 1111
print(num1, "-", check_dominance(num1))
print(num2, "-", check_dominance(num2))
print(num3, "-", check_dominance(num3))
```

```python
#program to convert a column matrix to row matrix col_matrix: [1, 2, 3, 4]
def col_to_row_matrix(col_matrix):
    row_matrix = [[element] for element in col_matrix]
    return row_matrix

col_matrix = [1, 2, 3, 4]
row_matrix = col_to_row_matrix(col_matrix)
print("Column matrix:", col_matrix)
print("Row matrix:", row_matrix)
```

```python
'''write a program to calculate sum of
 prime numbers within a given range.'''
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

def sum_of_primes(start, end):
    prime_sum = 0
    for num in range(start, end + 1):
        if is_prime(num):
            prime_sum += num
    return prime_sum
start_range = 1
end_range = 20
result = sum_of_primes(start_range, end_range)
print("Sum of prime numbers within the range", start_range, "to", end_range, "is:", result)
```
