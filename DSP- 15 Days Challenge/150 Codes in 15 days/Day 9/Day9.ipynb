```python
'''Write a program to take user-id
and password, if it doesn't match
to default data raise an error
displaying message:
'Invalid-Credentials'.
If credentials are valid display
message 'Home-page'.'''
def login(user_id, password):
    default_user_id = "user123"
    default_password = "pass123"
    if user_id == default_user_id and password == default_password:
        print("Home-page")
    else:
        raise ValueError("Invalid-Credentials")

try:
    user_id = input("Enter user ID: ")
    password = input("Enter password: ")
    login(user_id, password)
except ValueError as e:
    print(e)
```

```python
'''Write a program to create class Ticket:
Construct object by taking keyword argument.
Create a method to display passenger-details.
Define a method to show total fare.
Concession of 10% is given for senior citizen.
Display passenger details along fair and grand total.'''
class Ticket:
    def __init__(self, **kwargs):
        self.passenger_details = kwargs
        self.base_fare = 1000  # Assuming base fare for ticket

    def display_passenger_details(self):
        print("Passenger Details:")
        for key, value in self.passenger_details.items():
            print(f"{key}: {value}")

    def calculate_total_fare(self):
        age = self.passenger_details.get('age', 0)
        if age >= 60:  # Senior citizen concession
            total_fare = 0.9 * self.base_fare
        else:
            total_fare = self.base_fare
        return total_fare

    def display_ticket_details(self):
        print("\nTicket Details:")
        self.display_passenger_details()
        total_fare = self.calculate_total_fare()
        print(f"Base Fare: {self.base_fare}")
        print(f"Total Fare: {total_fare}")

ticket_details = {'name': 'John Doe', 'age': 65, 'gender': 'Male'}
ticket = Ticket(**ticket_details)
ticket.display_ticket_details()
```

```python
'''3. Write a program to ask for strong password until user give strong password by using do-while loop. A strong password should contain letter, number and special symbol.'''
import re

def is_strong_password(password):
  return re.search(r'[a-zA-Z]', password) and re.search(r'[0-9]', password) and re.search(r'[^a-zA-Z0-9]', password)

password = input("Enter a strong password: ")

while not is_strong_password(password):
  password = input("Password is not strong enough. Enter a strong password: ")

print("You have entered a strong password.")
```

```python
#Write a program to find median of list
def median(lst):

  lst.sort()
  middle = len(lst) // 2
  if len(lst) % 2 == 0:
    return (lst[middle-1] + lst[middle]) / 2
  else:
    return lst[middle]


lst = [10, 20, 30, 40, 50,60]
print(median(lst))
```

```python
#Write a function to take 4-bit binary number as parameter
def binary_to_decimal():

  binary_number = input("Enter a 4-bit binary number: ")

  if not (len(binary_number) == 4 and all(c in ['0', '1'] for c in binary_number)):
    raise ValueError("Invalid binary number.")

  decimal_number = 0
  for i, bit in enumerate(reversed(binary_number)):
    if bit == '1':
      decimal_number += 2**i
  print("Decimal equivalent:", decimal_number)

binary_to_decimal()
```

```python
# program to find cubes of numbers of list using map () function
def cube(n):
  return n**3

numbers = [1, 2, 3, 4, 5]
cubed_numbers = map(cube, numbers)
cubed_numbers_list = list(cubed_numbers)
print(cubed_numbers_list)
```

```python
'''. write a program to input a list which contains strings start with vowels and consonants. Remove all the string starts with vowels.'''
def remove_vowel_strings(string_list):
  vowels = {'a', 'e', 'i', 'o', 'u'}
  consonant_strings = [string for string in string_list if string[0].lower() not in vowels]

  return consonant_strings
string_list = ["apple", "banana", "car", "dog", "elephant"]
consonant_strings = remove_vowel_strings(string_list)
print(consonant_strings)  # Output: ['car', 'dog']
```

```python
#Write a program to transpose of two-dimensionalÂ matrix.
def transpose_matrix(matrix):
  rows = len(matrix)
  cols = len(matrix[0])
  transposed_matrix = [[0 for _ in range(rows)] for _ in range(cols)]
  for i in range(rows):
    for j in range(cols):
      transposed_matrix[j][i] = matrix[i][j]
  return transposed_matrix

matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
transposed_matrix = transpose_matrix(matrix)

print("Original matrix:")
for row in matrix:
  print(row)

print("\nTransposed matrix:")
for row in transposed_matrix:
  print(row)
```
