```python
#find index in a list
my_list = [1, 3, 5, 7, 9]
element_to_find = 5
index = -1
for i in range(len(my_list)):
    if my_list[i] == element_to_find:
        index = i
        break
if index != -1:
    print(f"The element {element_to_find} is found at index {index}.")
else:
    print(f"The element {element_to_find} is not in the list.")
```

```python
#program to find fibonacci series
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def print_fibonacci_series(count):
    if count <= 0:
        print("Please enter a positive integer.")
    else:
        print("Fibonacci series:")
        for i in range(count):
            print(fibonacci(i), end=" ")
count = int(input("Enter the number of Fibonacci numbers to print: "))
print_fibonacci_series(count)
```

```python
#creating class and performing tasks
class Student:
    def __init__(self, name, age, marks1, marks2, marks3):
        self.name = name
        self.age = age
        self.marks1 = marks1
        self.marks2 = marks2
        self.marks3 = marks3
student1 = Student("Asra", 20, 85, 90, 78)
print("Name:", student1.name)
print("Age:", student1.age)
print("Marks 1:", student1.marks1)
print("Marks 2:", student1.marks2)
print("Marks 3:", student1.marks3)
```

```python
#to find max element
def find_max(num1, num2, num3):
    max_num = max(num1, num2, num3)
    return max_num
number1 = int(input("Enter the first number: "))
number2 = int(input("Enter the second number: "))
number3 = int(input("Enter the third number: "))
maximum = find_max(number1, number2, number3)
print(f"The maximum of {number1}, {number2}, and {number3} is: {maximum}")
```

```python
#max element in the list
def find_max_element(lst):
    if not lst:
        return None
    max_element = lst[0]
    for element in lst[1:]:
        if element > max_element:
            max_element = element
    return max_element
my_list = [5, 9, 3, 8, 2, 7]
maximum = find_max_element(my_list)
if maximum is not None:
    print(f"The maximum element in the list is: {maximum}")
else:
    print("The list is empty.")
```

```python
#factorial of number
def factorial(n):
    return 1 if n == 0 else n * factorial(n - 1)
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))
sum_of_factorials = factorial(num1) + factorial(num2)
print(f"The sum of factorials of {num1} and {num2} is: {sum_of_factorials}")
```

```python
#find if it's a mail or not
def gmail(email):
    return email.endswith("@gmail.com")
email1 = "example@gmail.com"
email2 = "test@example.com"
result1 = gmail(email1)
result2 = gmail(email2)
print(f"{email1} have '@gmail.com' at the end {result1}")
print(f"{email2} have '@gmail.com' at the end {result2}")
```

```python
'''Define a function that accepts a value. if the value is >=200. print “You cannot increase”. Else increase the value by 5 and print the updated value.'''
def increase_value(value):
    if value >= 200:
        print("You cannot increase")
    else:
        updated_value = value + 5
        print("Updated value:", updated_value)
input_value = int(input("Enter a value: "))
increase_value(input_value)
```
