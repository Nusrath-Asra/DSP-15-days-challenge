```python
#find the greatest among 3
n1=int(input("Enter number 1: "))
n2=int(input("Enter number 2: "))
n3=int(input("Enter number 3: "))
if n1>n2 and n1>n3:
   print(n1,"is greater than",n2,"and",n3)
elif n2>n1 and n2>n3:
   print(n2, "is greater than",n1,"and",n3)
else:
   print(n3, "is greater than",n1,"and",n2)
```

```python
#swapping of numbers
V1=int(input("Enter a value: "))
V2=int(input("Enter another value: "))
V1,V2=V2,V1
print("Value 1 after swapping: ",V1)
print("Value 2 after swapping: ",V2)
```

```python
#factorial of a num
def fact(n):
    if n==0:
       return 1
    else:
       return n*fact(n-1)
n=int(input("Enter number: "))
print("Factorial is: ",fact(n))
```

```python
#finding a leap year
year = int(input("Enter a year: "))
leap_year = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

if leap_year:
    print(year, "is a leap year.")
else:
    print(year, "is not a leap year.")
```

```python
#finding palindrome
string=input().lower()
rev_string=string[::-1]
if string==rev_string:
   print(string," is palindrome")
else:
   print(string,"is not palindrome")
```

```python
#counting vowels and consonents in a string
string=input("Enter a string: ").lower()
counter=0
c=0
for i in string:
    if i=='a' or i =='e' or i== 'i' or i== 'o' or i== 'u':
       counter=counter+1
    else:
       c=c+1
print("Vowels are: ",counter)
print("Conso are: ",c)
```
