```python
class BankAccount:
    def __init__(self, user_name, user_id, pin):
        self.__user_name = user_name
        self.__user_id = user_id
        self.__pin = pin
        self.__balance = 0

    def __deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Amount deposited: {amount}")
            print(f"New balance: {self.__balance}")
        else:
            print("Invalid deposit amount.")

    def __withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount
            print(f"Amount withdrawn: {amount}")
            print(f"Remaining balance: {self.__balance}")
        else:
            print("Insufficient funds or invalid withdrawal amount.")

    def transaction(self, pin, amount, transaction_type):
        if pin == self.__pin:
            if transaction_type == "deposit":
                self.__deposit(amount)
            elif transaction_type == "withdraw":
                self.__withdraw(amount)
            else:
                print("Invalid transaction type.")
        else:
            print("Incorrect PIN.")
account = BankAccount("Asra", "123456789", "1111")
account.transaction("1111", 500, "deposit")
account.transaction("1111", 200, "withdraw")
account.transaction("1111", 1000, "withdraw")
```

```python
class HouseSearch:
    def __init__(self, data):
        self.data = data

    def search(self, budget, house_type, is_furnished):
        results = []
        for house_id, attributes in self.data.items():
            rent, h_type, furnished = attributes
            if h_type == house_type and furnished == is_furnished and rent <= budget:
                results.append(house_id)
        return results
houses = {
    "house1": [1000, "1bhk", True],
    "house2": [1500, "2bhk", False],
    "house3": [1200, "3bhk", True],
    "house4": [900, "1bhk", False],
    "house5": [2000, "2bhk", True],
}

search_engine = HouseSearch(houses)

budget = 1300
house_type = "2bhk"
is_furnished = True

search_results = search_engine.search(budget, house_type, is_furnished)
print("Search Results:")
for result in search_results:
    print(result)
```

```python
'''a program to take a number from
user and check whether it is part of
 Fibonacci sequence or not.'''
def is_perfect_square(n):
    sqrt = int(n ** 0.5)
    return sqrt * sqrt == n

def is_fibonacci(n):
    return is_perfect_square(5 * n * n + 4) or is_perfect_square(5 * n * n - 4)

number = int(input("Enter a number: "))
if is_fibonacci(number):
    print(number, "is part of the Fibonacci sequence.")
else:
    print(number, "is not part of the Fibonacci sequence.")
```

```python
'''function to take a list as
parameter and return a list by sorting list where even numbers first and odd
numbers comes after even numbers.'''
def sort_even_odd(input_list):
    even_numbers = [num for num in input_list if num % 2 == 0]
    odd_numbers = [num for num in input_list if num % 2 != 0]
    sorted_list = sorted(even_numbers) + sorted(odd_numbers)
    return sorted_list
actual_list = [9, 2, 3, 7, 1, 4, 5, 8, 0, 6]
sorted_list = sort_even_odd(actual_list)
print("Sorted list:", sorted_list)
```

```python
#program to calculate total fine in a month.
def calculate_fine(ticket_list):
    total_tickets = len(ticket_list)
    total_fine = 0
    for ticket in ticket_list:
        if ticket > 50:
            total_fine += 50
    return total_tickets, total_fine

ticket_list = [92, 44, 55, 77, 82]
total_tickets, total_fine = calculate_fine(ticket_list)
print("Total tickets raised:", total_tickets)
print("Total fine charged:", total_fine)
```

```python
'''program to create list_extenstion
 class inheriting list class and define a method that return all the indexes of element as list if there no element
presents in list return empty list.'''

class list_extension(list):
    def indexes(self, element):
        indexes_list = []
        for i in range(len(self)):
            if self[i] == element:
                indexes_list.append(i)
        return indexes_list

lst = [121, 115, 89, 54, 121, 110, 33, 92, 44, 67]
obj = list_extension(lst)
print("Indexes of 121:", obj.indexes(121))
obj.sort()
print("Sorted list:", obj)

obj.reverse()
print("Reversed list:", obj)

obj.append(100)
print("Appended list:", obj)
```

```python
#program to play â€œrock-paper-scissor
import random
def play_game():
    choices = ["rock", "paper", "scissors"]

    player_choice = input("Enter your choice (rock, paper, scissors): ").lower()
    while player_choice not in choices:
        print("Invalid choice. Please enter rock, paper, or scissors.")
        player_choice = input("Enter your choice (rock, paper, scissors): ").lower()

    computer_choice = random.choice(choices)

    print("Computer chose:", computer_choice)
    print("Player chose:", player_choice)

    if player_choice == computer_choice:
        print("It's a tie!")
    elif (player_choice == "rock" and computer_choice == "scissors") or \
         (player_choice == "paper" and computer_choice == "rock") or \
         (player_choice == "scissors" and computer_choice == "paper"):
        print("Player wins!")
    else:
        print("Computer wins!")
play_game()
```

```python

#program to check if sum of any two numbers in list make the number taken from console.
def sort_dict_by_values(input_dict):
    return {k: v for k, v in sorted(input_dict.items(), key=lambda item: item[1])}

def check_sum(numbers_list, target):
    for i in range(len(numbers_list)):
        for j in range(i+1, len(numbers_list)):
            if numbers_list[i] + numbers_list[j] == target:
                return numbers_list[i], numbers_list[j]
    return None
input_dict = {'a': 'apple', 'b': 'banana', 'c': 'cat', 'd': 'dog'}
sorted_dict = sort_dict_by_values(input_dict)
print("Sorted dictionary by values:", sorted_dict)

numbers_list = [23, 45, 72, 19, 35]
target = int(input("Enter a number: "))
pair = check_sum(numbers_list, target)
if pair:
    print(f"Sum ({pair[0]}, {pair[1]}) from the list makes {target}.")
else:
    print("No pair of numbers in the list makes the given number.")
```
